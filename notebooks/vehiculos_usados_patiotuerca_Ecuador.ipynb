### 01. Exploracion del Dataset

```python
import pandas as pd

# Definición de códigos ANSI (Solo texto)
VERDE = '\033[92m'
ROJO = '\033[91m'
NEGRITA = '\033[1m'
FIN = '\033[0m'
path = '../data/patiotuerca2023-02-10.csv'
df_vehiculos = pd.read_csv(path)
print(f"{VERDE}{NEGRITA}ESTADO: IMPORTACION EXITOSA{FIN}")
```

## 01.1 Filas columnas y header

```python
#Columnas y filas totales

print(f'{ROJO}{NEGRITA}El numero de filas y columnas : {FIN}{df_vehiculos.shape}')
print(50*'-')
#Visualizacion Header
print(f'{ROJO}{NEGRITA}Impresion del header {FIN}\n{df_vehiculos.head(5)}')
#Ultimas filas
```

## 01.2 Colas e info

```python
print(f'{ROJO}{NEGRITA}Ultimas 5 filas :{FIN}\n {df_vehiculos.tail(2)}')
print(50*'-')
print(f'{ROJO}{NEGRITA}Impresion de la info :{FIN} \n{df_vehiculos.info()}')
```

## 01.3 Descripcion y descripcion categorica

```python
print(f'{ROJO}{NEGRITA}Impresion de la descripcion :{FIN}\n{df_vehiculos.describe()}')
print(50*'-')
print(f'{ROJO}{NEGRITA}Impresion de la descripcion todo:{FIN}\n{df_vehiculos.describe(include="all")}')
```

## 01.4 Nombres de columnas, tipo de columna y valores unicos

```python
print(f'{ROJO}{NEGRITA}Impresion lista de columnas :{FIN}\n{df_vehiculos.columns.to_list()}')
print(50*'-')
print(f'{ROJO}{NEGRITA}Impresion de tipo de columnas :{FIN}\n{df_vehiculos.dtypes}')
print(50*'-')
```
